```python
import pandas as pd
path_in="/home/noura369/Documents/Basic_NOURA2/ABC/titanic-passengers.csv"
df =pd.read_csv(path_in,sep=';')
df.isnull().sum()
#df.head()
```

```python
df.head()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(df['Survived'])
plt.show()
```

```python
sns.boxplot(df['Sex'], df["Age"])
```

```python
sns.boxplot(df['Sex'], df["Age"], df["Survived"])
plt.show()
```

```python
sns.barplot(df['Pclass'], df['Fare'], hue = df["Survived"])
plt.show()
```

```python
sns.barplot(df['Pclass'], df['Fare'], hue = df["Sex"])
plt.show()
```

```python
sns.barplot(df['Pclass'], df['Fare'], hue = df["Age"])
plt.show()
```

```python
sns.barplot(df['Pclass'], df['Age'])
plt.show()
```

```python
sns.distplot(df['Pclass'])
plt.show()
```

```python
sns.distplot(df['Age'],hist=False)
plt.show()
```

```python
sns.violinplot(x="Survived", y="Age", hue='Sex', data=df)
plt.show()
```

```python
plt.subplots(figsize=(10, 10))
sns.heatmap(df.corr(), cmap = "Greens", annot=True, fmt=".2f")
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.scatterplot(x="Pclass", y="Fare", hue="Age", data=df)
plt.show()
```
